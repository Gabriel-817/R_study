### 문제정의
농구 선수의 경기 기록을 바탕으로 그 선수의 포지션 예측

### 데이터 수집

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/My Drive/업종별 분석/data/basketball_stat.csv")
```

```python
df.head()
# Player	
# Pos (SG, C)	
# 3P(3점슛)	
# 2P(2점슛)	
# TRB(리바운드)	
# AST(어시스트)	
# STL(스틸)	
# BLK(블로킹)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 스틸, 2접슛 데이터 시각화
sns.lmplot('STL', '2P', data=df, fit_reg=False, # x 축, y 축, 라인 없음
          scatter_kws={"s": 150}, # 좌표 상의 점 크기
          markers=["o", "x"],
          hue="Pos") #예측값
plt.title("STL and 2P in 2d plane")
```

```python
# 스틸, 어시스트 데이터 시각화
sns.lmplot('STL', 'AST', data=df, fit_reg=False, # x 축, y 축, 라인 없음
          scatter_kws={"s": 150}, # 좌표 상의 점 크기
          markers=["o", "x"],
          hue="Pos") #예측값
plt.title("STL and AST in 2d plane")
```
