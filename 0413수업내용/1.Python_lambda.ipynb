## 람다식의 기초
### 익명 함수 만들기

### 기본 함수 정의

```python
# 예: x^2을 출력하는 함수 pow1(x)
def pow1(x):
    return x ** 2
```

### 익명 함수를 사용하여 코드를 간단히 정의
 - anonymous function  
 - lambda (람다식)  
람다식 구조  
lambda 인수 : 반환 값

```python
# pow1(x) 와 동일한 람다식
pow2 = lambda x : x ** 2
```

```python
pow1(10)
pow2(10)
```

### 인수 a에 대한 다음 함수를 작성 a=4일때의 반환값 출력
- def 이용하여 2a^2 - 3a + 1
- lambda 이용 2a^2 - 3a + 1

```python
a=4
def func1(x):
    return 2 * x**2 - 3 * x + 1

func2 = lambda x : 2 * x**2 - 3 * x + 1

print(func1(a))
print(func2(a))
```

### 람다식으로 다변수 함수 정의

```python
add1 = lambda x, y: x + y

print(add1(1,2))
```

## 인수(x,y)에 대한 함수를 작성하고 (x,y,z) = (5, 6, 2)일때의 반환값을 각각 출력
- def를 이용하여 xy + z을 출력하는 함수 func3  
- lambda를 이용하여 xy + z를 출력하는 함수 func4


```python
# 대입할 인수 x, y, z 선언
x = 5
y = 6
z = 2

# def를 이용한 함수
def func3(x, y, z):
    return x * y + z

# lambda를 이용한 함수
func4 = lambda x, y, z: x * y + z

print(func3(x, y, z))
print(func4(x, y, z))
```

### if 조건문에서 삼항 연산자를 이용

```python
def lower_threel(x):
    if x < 3:
        return x *2
    else:
        return x/3 + 5
```

```python
lower_threel = lambda x: x * 2 if x < 3 else x/3 + 5
```

### lambda를 이용하여 함수를 작성하고 인수 a에 대한 결과 출력
- a가 10이상 30미만이면 a^2 - 40a의계산값을, 이외에는 50을 출력

```python
# 대입 할 인수
a1 = 13
a2 = 32

# lambda를 이용한 func5
func5 = lambda x: x**2 - 40 * x if x >= 10 and x < 30 else 50

#반환값 출력
print(func5(a1))
print(func5(a2))
```

### "My name is OOO" 문자열 self_data 에서 "OOO" 부분만 추축

```python
# 자기소개 문자열
self_data = "My name is baniota"

# 분리 해서 리스트
word_list = self_data.split(" ")

# "baniota" 을 출력
print(word_list[len(word_list)-1])
```

### 여러 기호의 구분자로 문자열 분할

```python
import re
test_sentence = "this,is a.test,sentence."
# ","와 " " 그리고 "." 으로 구분자 지정 필요
re.split("[, .]", test_sentence)
```

### "년/월/일 시:분" 구조를 갖는 문자열 time_data를 "월"과 "시"로 분할해서 출력

```python
import re
time_data = "2020/4/1_22:15"
time_list = re.split("[/_:]", time_data)

# "월"과 "시"를 출력
print(time_list[1])
print(time_list[3])
```

### 리스트 내포

```python
# 리스트 내포로 각 요소의 절대값을 취하시오
a = [1, -2, 3, -4, 5]
# def abs_fun(x):
#     for i in a:
#         abs(a[i])
print([abs(x) for x in a])
```

```python
# map 으로 리스트 
a = [1, -2, 3, -4, 5]
print(list(map(abs, a)))
```

```python
# 분
minute_data = [30, 155, 180, 74, 11, 60, 82]

# 시, 분으로 계산 lambda식
h_m_split = lambda x:[x // 60, x % 60]

# 리스트 내포를 이용 배열 작성
h_m_data = [h_m_split(x) for x in minute_data]

print(h_m_data)
```
